plot canopy height

```python
is2_file = "sample_data/processed_ATL08_20200104180343_01380602_006_01.h5"
reader = ipx.Read(is2_file)
reader.vars.append(beam_list=['gt1l'], var_list=['h_canopy', "latitude", "longitude"])
ds = reader.load()
ds
ds.plot.scatter(x="longitude", y="latitude", hue="h_canopy", vmin=-0, vmax=100).figure
```
